```python
a="The quick brown fox jumps over the lazy dog"
a=a.lower()
b=list(a)
#print(b)
s=set(b)
print(s)
c=0
for i in s:
    if i.isalpha()==True:
        c=c+1
if c==26:
    print("It is palingram")  # checks if the string contains unique 26 letters or nor
else:
    print("It is not palingram")
```

```python
A={3,4,5,6}
B={3,4}
l=len(B)
c=0
for i in B:
    for j in A:
        if i==j:
            c=c+1
if c==l:
    print("B is subset of A")
else:
    print("B is not subset of A")
        
```
