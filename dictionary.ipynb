```python
l=[1,2,3,4,5]
k=[2,"Hello",3.5,"H",6]
d={}
for i in range(len(l)):      #map two lists into dictionary
    d[l[i]]=k[i]
print(d)
    
```

```python
d={
    7:"sai",
    6:"krishna",
    1:"sumanth",
    2:"pavan"
}
d1={}
l=list(d.keys())
l.sort()
for i in l:           #sort dictionary by keys
    d1[i]=d[i]
print(d1)

```

```python
temp={
    "chennai":28,
    "Banglore":12,
    "Nellore":34,
    "Tirupati":18,
    "kakinada":15
}
t=list(temp.values()) # sort the dictionary values and result is in dictionary type it  
                           #is converted into list using list conversion.
t.sort()
for i in range(0,3):
    for j in temp:
        if temp[j]==t[i]:
            print(j)
```

```python
a="Fear leads to Anger and Anger leads to Fear"
a.lower()
f=a.split()
d={}
for i in f:
    if d.get(i)==None:     # count the no of elements present in the dictionary
        d[i]=1
    else:
        d[i]=d[i]+1
print(d)
```
