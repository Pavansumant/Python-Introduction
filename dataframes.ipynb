# Ranking using rank() function

```python
import pandas as pd
d={
    "country":["india","America","Japan","Us","Singapore"],
    "Gold":[99,34,56,23,None]
}
```

```python
df=pd.DataFrame(d)
df
```

**We are given rank based on gold.**

```python
df["Rank"]=df["Gold"].rank(ascending=False,na_option="bottom")  #  rank is specified in descending order and nan value is replaced by hi 
df.sort_values("Rank",inplace=True)
df
```

If we have two same values,it takes average of position by default.

```python
d={
    "country":["india","America","Japan","Us","Singapore"],
    "Gold":[99,34,19,34,None]
}
```

```python
df=pd.DataFrame(d)
df
```

```python
df["Rank"]=df["Gold"].rank()   
df.sort_values('Rank',ascending=False,inplace=True)
df
```

```python
df["Rank"]=df["Gold"].rank(ascending=False,na_option="bottom")  #  rank is specified in descending order and nan value is replaced by hi 
df
```

**We use method attribute to specify the min or max value instead of taking avg function if we have same values.**

```python
df["Rank"]=df["Gold"].rank(ascending=False,na_option="bottom",method="min")  #  rank is specified in descending order and nan value is replaced by hi 
df
```
