```python
n=int(input("Enter a item:"))
l=[1,4,1,5,6]
c=0
for i in l:
    if i==n:
        c=c+1   # checks how many items a particular element present in the list
print(c)
```

```python
n=int(input("Enter a item:"))
l=[1,4,1,5,6]
for i in l:
    if i==n:
        l.remove(n)  # removes particular element in the list
print(l)
```

```python
l=[1,9,8,6]
s=[2,4,6,8]
for i in l:
    for j in s:
        if i==j:
            print(i)  # to find common elements present in the list
            break
    
```

```python
l=[1,4,9,2,1,6,2]
for i in l:
    c=0
    for j in l:
        if i==j:
            c=c+1
            if c>1:
                l.remove(j)  # to remove duplicate elements in the list
print(l)
```
