```python
def sol(name):
    c=0
    for i in name:
        if i.isupper():
            c=c+1
        return c
```

```python
sol("PAVAn")
```

```python
def sol(x,y):
    a=x+y
    b=x-y
    c=x*y
    d=x//y
    return a,b,c,d
    
```

```python
add,sub,mul,div=sol(3,4)
print(add,sub,mul,div)
```
