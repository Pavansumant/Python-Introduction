**List Vs numpy**

numpy is faster than list.
it contains homogeneous data.Memort alloc is optimised.
computation is faster as it built  on c in backend  as execution of c is faster compared to python.


```python
import numpy as np
import time
```

```python
intial_time=time.time()    #using list
a=0
for i in range(1000000):
    a=a+i
print(a)
final_time=time.time()
diff=final_time-intial_time
print(diff)
```

```python
intial_time=time.time()    #using numpy
a=np.arange(1000000)
s=np.sum(a)
print(s)
final_time=time.time()
diff1=final_time-intial_time
print(diff1)
```

```python
diff/diff1   # here numpy is 23 times faster than list
```

In numpy,the mathemeatical operation are element wise,hence these arrays must have same size/shape.
But sometimes numpy can itself convert different shaped array of same shape using concept of broadcasting.

**Type1 Broadcasting- Rows are repeated**

**Type2 Broadcasting- Columns are repeated**

**Type3 Broadcasting- Rows and Columns are repeated**


```python
x=np.arange(18).reshape(3,6)
x
```

```python
y=np.array([1,2,3,4,5,6])
y
```

```python
x+y   #Type 1 broadcasting rows are repeated.
```

```python
y=np.array([1,2,3]).reshape(3,1)
y
```

```python
x+y     # type 2 broadcasting columns are added.
```

```python
x=np.array([1,2,3])
y=np.array([[1],[2],[3]])
```

```python
x+y   #type 3 broadcasting rows and columns are added.
```

```python
x=np.array([[1,2,3],[4,5,6]])
x
```

```python
x[0,1]
```
