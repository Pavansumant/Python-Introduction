# Find duplicates **using duplicated() method**.


**It is used to find dupicated records.**

```python
a={
    "student":["pavan","sumanth","pavan","sai","pavan"],
    "age":[18,21,22,12,18],
    "marks":[25,22,41,22,25]
}
```

```python
df=pd.DataFrame(a)
df
```

```python
mask=df.duplicated()
mask# it returns true when whole record is repeated otherwise false. Here sth record is repeated same as first oone so true.
```

```python
df[mask]
```

```python
mask=df["student"].duplicated()
mask                           # it is different from previous one.it finds out duplicates only in name column and returns true if duplicate is found.
```

```python
df[mask]   # here 2 and 4 uplicated in names columns so those records are returned.
```

```python
mask=df.duplicated(subset=["student","age","marks"])  # iit returns duplkicate record based on the name,marks,age column
mask
```

```python
df[mask]
```

**Treating Duplicates using drop_duplicates() method**

```python
import pandas as pd 
a={
    "student":["pavan","sumanth","pavan","sai","pavan"],
    "age":[18,21,22,12,18],
    "marks":[25,22,41,22,25]
}
df=pd.DataFrame(a)
df
```

```python
mask=df["student"].drop_duplicates()
mask# it drops duplicates in the names column and remove the duplicates.
```

```python
df.drop_duplicates(subset=["student","age","marks"])  # only 5th row is repeated it is same as ist row so it is removed but 3rd row is not removed beacuse only name is repeated but not marks and age.we specified all three conditions.so it is not removed.
```

#  Difference between Mask filtering and where condition
 
**Mask filtering returns only true records.
but where condition returns true records as well as it  return records that are false  but it will fill those details with null values.**

```python
a.where(a["Location"]=="Bangalore/Bengaluru")  # it returns true records and also return records that do not have loaction specified but it wiill those details with null values.
```

```python
import numpy as np
a=np.random.randint(10,200,(4,4))
a
```
