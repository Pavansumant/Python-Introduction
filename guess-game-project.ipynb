```python
import random
```

```python
words=["pavan","sumanth","sai","vinay"]
word=random.choice(words)
print(word)
```

```python
def word_so_far(word,guesses):
  guess=''
  for l in word:
      if l in guesses:
        guess=guess+l
      else:
        guess=guess+'-'
  return guess
```

```python
chances=len(word)*2
guessed=[]
while True:
  if chances!=0:
    print("Guess a word")
    print("you have left with",chances," chances in hand")
    print(word_so_far(word,guessed))
    print("Guess a letter")
    s=input()
    if s not in guessed:
      guessed.append(s)
    if word_so_far(word,guessed)==word:
      print("congratulations you have got right word",word)
      break
    else:
      chances=chances-1
      if s in word:
        print("you have correctly guessed")
      else:
        print("you have wrongly guessed")
  else:
    print("you chances are over")
    break
```
